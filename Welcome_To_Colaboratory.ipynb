```
#But to really understand this we would need to look at the total market cap of the company, not just the stock price. 
#Unfortunately our current data doesn't have that information of total units of stock present.
# But what we can do as a simple calcualtion to try to represent total money traded would be to multply the Volume column by the Open price. This still isn't the actual Market Cap, its just a visual representation of the total amount of money being traded around using the time series.
```

```
FORD['Total Traded'] = FORD['Open']*FORD['Volume']
GM['Total Traded'] = GM['Open']*GM['Volume']
TM['Total Traded'] = TM['Open']*TM['Volume']

VWAGY['Total Traded'] = VWAGY['Open']*VWAGY['Volume']
RNSDF['Total Traded'] = RNSDF['Open']*RNSDF['Volume']
PEUGF['Total Traded'] = PEUGF['Open']*PEUGF['Volume']
```

```

FORD['Total Traded'].plot(label='Ford',figsize=(16,10))
GM['Total Traded'].plot(label='GM')
TM['Total Traded'].plot(label='TM')

VWAGY['Total Traded'].plot(label='VWAGY')
RNSDF['Total Traded'].plot(label='RNSDF')
PEUGF['Total Traded'].plot(label='PEUGF')

plt.legend()
plt.ylabel('Total Traded')
```
